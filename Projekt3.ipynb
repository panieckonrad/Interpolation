```python
import matplotlib.pyplot as plt
import pandas as pd
import math as math
import numpy as np
```

```python
dataset = "MountEverest.csv"
df = pd.read_csv("2018_paths/"+dataset)
df.head(n=512)
```

```python
npmatrix = df.values # z dataframe robimy matrix
del df# usuwamy niepotrzebny df
```

```python
print(npmatrix)
```

```python
skipNumber = 54
npxcopy = npmatrix[0:,0] # wyluskujemy wektor x
npycopy = npmatrix[0:,1] # wyluskujemy wektor y
npx = npxcopy[::skipNumber]
npy = npycopy[::skipNumber]
```

```python
def base(x,i):# x = zmienna, i = indeks
    baseReturn = 1.0
    for j in range(0,npx.size):
    #for j in range(0,size):
        if j == i:
            continue
        numerator = x - npx[j]
        denominator = npx[i]-npx[j]
        baseReturn = baseReturn*(numerator/denominator)
    return baseReturn
```

```python
def F(x):
    sum = 0
    for i in range (0,npx.size):
    #for i in range (0,size):
        sum+= npy[i] * base(x,i)
    return sum
        
```

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(npx,npy,'ro');
```

```python
ymin = np.amin(npycopy)  - 40
ymax = np.amax(npycopy)  + 30
fig = plt.figure()
ax = plt.axes()
plt.ylim((ymin, ymax))
ax.plot(npxcopy,npycopy)
ax.plot(npx,npy,'ro', markersize = 3)
ax.plot(npxcopy,F(npxcopy));
```
